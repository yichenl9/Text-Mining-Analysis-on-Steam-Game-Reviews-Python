```python
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        
def get_topic_words(vectorizer, lda_model, n_words):
    keywords = np.array(vectorizer.get_feature_names())
    topic_words = []
    for topic_weights in lda_model.components_:
        top_word_locs = (-topic_weights).argsort()[:n_words]
        topic_words.append(keywords.take(top_word_locs).tolist())
    return topic_words
```

```python
emotions_elements = []
def NRC_test(text):   
    for i in range(len(text)):         
        emotion = NRCLex(text[i])        
        for a_tuple in emotion.affect_list:
            emotions_elements.append(a_tuple)
    fear_reviews = emotions_elements.count('fear')
    anger_reviews = emotions_elements.count('anger')
    sadness_reviews = emotions_elements.count('sadness')
    disgust_reviews = emotions_elements.count('disgust')
    trust_reviews = emotions_elements.count('trust')
    anticipation_reviews = emotions_elements.count('anticipation')
    surprise_reviews = emotions_elements.count('surprise')
    joy_reviews = emotions_elements.count('joy')
    eight_emotions_reviews =  fear_reviews +  anger_reviews + sadness_reviews +  disgust_reviews + trust_reviews +  anticipation_reviews +  surprise_reviews + joy_reviews
    print("fear reviews:",fear_reviews,'\n',round(fear_reviews/eight_emotions_reviews*100,3),'%')
    print("anger reviews:",anger_reviews,'\n',round(anger_reviews/eight_emotions_reviews*100,3),'%')
    print("sadness reviews:",sadness_reviews,'\n',round(sadness_reviews/eight_emotions_reviews*100,3),'%')
    print("disgust reviews:",disgust_reviews,'\n',round(disgust_reviews/eight_emotions_reviews*100,3),'%')
    print("trust reviews:",trust_reviews,'\n',round(trust_reviews/eight_emotions_reviews*100,3),'%')
    print("anticipation reviews:",anticipation_reviews,'\n',round(anticipation_reviews/eight_emotions_reviews*100,3),'%')
    print("surprise reviews:",surprise_reviews,'\n',round(surprise_reviews/eight_emotions_reviews*100,3),'%')
    print("joy reviews:",joy_reviews,'\n',round(joy_reviews/eight_emotions_reviews*100,3),'%')
```
